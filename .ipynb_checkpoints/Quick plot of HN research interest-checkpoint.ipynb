```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
cum_num_articles = [5, 22, 46, 82, 148, 243, 357, 521, 710, 939, 1210, 1460, 1690, 1910, 2150, 2410, 2610, 2850, 3090, 3360, 3650, 3960, 4290, 4590, 4870, 5180, 5490, 5810, 6080, 6320, 6520, 6680, 6830, 6990, 7150, 7290]
cum_num_articles = np.array(cum_num_articles)
num_articles = cum_num_articles[1:] - cum_num_articles[:-1]
years = np.arange(1984, 2019)
```

```python
fig = plt.figure(figsize = (20,5))
plt.plot(years, num_articles)
plt.grid(True)
```
